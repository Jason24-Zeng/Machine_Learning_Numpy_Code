```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from scipy.stats import multivariate_normal
plt.style.use('seaborn')

```

```python
def create_sample(m_examples, mus, vars):
    #X = ([] for _ in range(len(mus)))
    total_X = np.array([])
    subsample = []
    for i in range(len(mus)):
        X = np.random.multivariate_normal(mus[i], np.diag(vars[i]), m_examples[i])
        subsample.append(X)
        if i == 0:
            total_X = X
        else:
            total_X = np.vstack((total_X, X))
    return total_X, subsample
```

```python
m_examples = [400, 600, 1000]
mus = [[0.5, 0.5], [5.5, 2.5], [1,7]]
vars = [[1, 3], [2, 2], [6, 2]]
```

```python
X, subsample = create_sample(m_examples, mus, vars)
X.shape
```

```python
plt.figure(figsize=(10,8))
plt.axis([-10, 15, -5, 15])
plt.scatter(subsample[0][:,0], subsample[0][:,1], s=5)
plt.scatter(subsample[1][:,0], subsample[1][:,1], s=5)
plt.scatter(subsample[2][:,0], subsample[2][:,1], s=5)
plt.show()
```
